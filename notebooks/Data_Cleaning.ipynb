```python
df['bathroom'].fillna(df['bathroom'].median(), inplace=True)
df['age'].fillna(df['age'].median(), inplace=True)

df.isnull().sum()
```

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df['price'] = df['price'].apply(lambda x: upper if x > upper else x)
```

```python
# For location
location_counts = df['location'].value_counts()
rare_locations = location_counts[location_counts < 10].index
df['location'] = df['location'].apply(lambda x: 'Other' if x in rare_locations else x)

# For builder
builder_counts = df['builder'].value_counts()
rare_builders = builder_counts[builder_counts < 10].index
df['builder'] = df['builder'].apply(lambda x: 'Other' if x in rare_builders else x)
```

```python
df.to_csv("new_cleaned_data.csv", index=False)
```
